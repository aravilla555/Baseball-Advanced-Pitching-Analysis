# More Visualizations

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
advanced = pd.read_csv("/Users/anishravilla/Desktop/Data Science Union/Clean Pitching Data.csv", float_precision = 'round_trip')
```

```python
advanced.head(20)
```

### Visualization 1: Line Plot

```python
#Average Strike out percentage grouped by Year
strike = advanced.groupby('Year')['SO%'].median()
sns.set_theme(style = 'dark', palette = 'husl')
sns.lineplot(data = strike).set(title = "Strike Out Percentage 2000-2020")
#Average Strike Out Percentage increased over the years
```

### Visualization 2: Bar Plot

```python
#Barplot comparing BA for each time over their duration
ba = advanced.groupby('Tm')['BA'].mean()
ba

plt.style.use('fivethirtyeight')
fig = plt.figure()
ax = fig.add_axes([0,0,2,1])
fig = ba.plot.bar(x = "Team")
fig = plt.xlabel("Team")
fig = plt.ylabel("Average BA from 2000-2021")
fig = plt.title("Average BA of MLB teams from 2020-2021")
```

### Visualization 3: Faceted Histogram

```python
plt.style.use('tableau-colorblind10')
fig, ax = plt.subplots(nrows = 2, ncols = 11, sharex = True, figsize = (80,10))
#Store years in a matrix that matches subplot arrangement
matrix = np.arange(2000,2022).reshape(2,11)

def plot_hist1(advanced, year, fig, ax, i, j):
    df = advanced.groupby('Year').get_group(year)
    df_0 = df.loc[df['playoff_status'] == 0, ]
    df_1 = df.loc[df['playoff_status'] == 1,]
    
    ax[i,j].hist(df_0['BA'], alpha = 0.5, label = ["Non-Playoff Teams"])
    ax[i,j].hist(df_1['BA'], alpha = 0.5, label = ["Playoff Teams"])
    ax[i,j].legend(loc='upper right')
    ax[i,j].axvline(x = df_0['BA'].mean(), ymin = 0, ymax = 6, color = "b", linestyle = "--" )
    ax[i,j].axvline(x = df_1['BA'].mean(), ymin = 0, ymax = 6, color = "red", linestyle = "--" )
    ax[i,j].set_title("Batting Averages for Playoff vs. NonPlayoff Teams for " + str(year))


for i in range(2):
    for j in range(11):
        plot_hist1(advanced, matrix[i][j], fig, ax, i, j)
```

### Histogram of Home Run Percentages

```python
fig1, plot = plt.subplots(1)
plot.hist((advanced[advanced['playoff_status'] == 0])['HR%'], alpha = 0.5, label = ["Non-Playoff Teams"])
plot.hist((advanced[advanced['playoff_status'] == 1])['HR%'], alpha = 0.5, label = ["Playoff Teams"])
plot.legend(loc='upper right')
plot.set_title("Distribution of Home Runs Allowed for Playoff vs Non-Playoff Teams")
```

### Visualization 4: Logistic Curve

```python
x = advanced['BB%']
y = advanced['playoff_status']

sns.set_theme(style = "dark", palette = "pastel")
sns.regplot(x, y, data = advanced, logistic = True, ci = None).set(title = "The Probability of Making Playoffs based on BB%")
```

### Visualization 5: Scatterplot

```python
sns.set_theme(style = "ticks", palette = "flare")
sns.jointplot(data = advanced, x = 'OBP', y = 'SLG', hue = 'playoff_status')
```

# Possible Machine Learning Methods

### Logistic Regression

Assumptions: 
- Response Variable is Binary (playoff_status)
- Independent Observations 
- No Multicollinearity if using Multiple Predictor Variables (check variance inflation factors)
- No extreme outliers in predictors
- Assume linear relationship between predictor and log(odds) of response variable
- Sample Size is Large

### Linear Discriminant Analysis:

Summary:
- Reduce dimensions (if using many parameters) so that we can visualize data on one set of axes, but also try to visualize distinction between the categories 
- Maximize the distance between the means of the categories for respective variables, and minimize the variance between each of the observations in each category
- Goal: Find the predictor that accounts for the most variation in the response variable

Assumptions:
- Assume predictors come from normal distribution
- If multiple predictors, assume each class has a **common covariance matrix**.
