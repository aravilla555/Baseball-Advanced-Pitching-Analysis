```python
import requests
from bs4 import BeautifulSoup 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Advanced Pitching Data Table

```python
pitch_stats_Request = requests.get("https://www.baseball-reference.com/leagues/majors/2021-advanced-pitching.shtml")
#Store team stats in requests object
```

```python
pitch_stats_soup = BeautifulSoup(pitch_stats_Request.text)
#create beautiful soup object
```

```python
#select nodes
pitchingTable = pitch_stats_soup.select('#teams_advanced_pitching .right , #teams_advanced_pitching th')
#creates an iterable that you need a for loop to access

pitching_table_elements = []
for element in pitchingTable:
    pitching_table_elements.append(element.text)
print(pitching_table_elements)

```

```python
pitchArr = np.array(pitching_table_elements)
pitchArr = pitchArr[5:len(pitchArr)]
#34 x 18 data frame 
pitch_table_arr = pitchArr.reshape([34, int(len(pitchArr)/ 34)])
pitch_table_arr[[32,33]] = pitch_table_arr[[33,32]]

pitch_table_arr



```

```python
#Generate Data Frame:
pitchData = pd.DataFrame(pitch_table_arr, columns = pitchArr[0:18])
pitchData = pitchData.drop([0,31,32,33], axis = 0)
pitchData = pitchData.drop(["EV", "HardH%"],axis = 1)
pitchData = pitchData.dropna(axis = 1)  

pitchData['Year'] = 2021

pitchData = pitchData.reindex(columns = ['Year', 'Tm', 'BA', 'OBP', 'SLG', 'OPS', 'BAbip', 'HR%', 'SO%', 'BB%', 'LD%', 'GB%','FB%','GB/FB', 'WPA', 'cWPA', 'RE24'])
pitchData

```

```python
def pitchYear(year):
    pitch_stats_Request = requests.get(f"https://www.baseball-reference.com/leagues/majors/{year}-advanced-pitching.shtml")
    pitch_stats_soup = BeautifulSoup(pitch_stats_Request.text)
    
    pitchingTable = pitch_stats_soup.select('#teams_advanced_pitching .right , #teams_advanced_pitching th')
    #creates an iterable that you need a for loop to access

    pitching_table_elements = []
    for element in pitchingTable:
        pitching_table_elements.append(element.text)
    
    pitchArr = np.array(pitching_table_elements)
    pitchArr = pitchArr[5:len(pitchArr)]
    #34 x 16 data frame 
    pitch_table_arr = pitchArr.reshape([34, int(len(pitchArr) / 34 )])
    pitch_table_arr[[32,33]] = pitch_table_arr[[33,32]]
    
    pitchData = pd.DataFrame(pitch_table_arr, columns = pitchArr[0:pitch_table_arr.shape[1]])
    pitchData = pitchData.drop([0, 31,32,33], axis = 0)
    
    #Drop extra data columns that aren't in older data frames
    if(pitch_table_arr.shape[1] == 18):   
        pitchData = pitchData.drop(["EV", "HardH%"],axis = 1)

    #Reindex columns with year at front
    pitchData['Year'] = year
    pitchData = pitchData.reindex(columns = ['Year', 'Tm', 'BA', 'OBP', 'SLG', 'OPS', 'BAbip', 'HR%', 'SO%', 'BB%', 'LD%', 'GB%','FB%','GB/FB', 'WPA', 'cWPA', 'RE24'])
    
    return pitchData

    

```

```python
df = pitchYear(2020)
df

```

```python
#Create conjoined data frame for advanced pitching
for year in range(2000, 2021):
    df = pitchYear(year)
    #print(year)
    pitchData = pd.concat([pitchData, df], axis = 0, ignore_index = False)

```

```python
#Sorting data uses those sort algorithms 
pitchData = pitchData.sort_values(by = ['Year', 'Tm'], ascending = [False, True])

pitchData


```

### Determine Playoff Status

```python
playoff_status = requests.get("https://www.baseball-reference.com/postseason/")
```

```python
#Read in postseason table elements
postseason_soup = BeautifulSoup(playoff_status.text)
postseason_Table = postseason_soup.select("#postseason_series a")

postseason_elements = []
for element in postseason_Table:
    postseason_elements.append(element.text)
postseason_elements = postseason_elements[0:540]

postseason_elements

```

```python
#Create dataframe for postseason
postseason_elements = np.array(postseason_elements).reshape(180,3)
postseason = pd.DataFrame(postseason_elements, columns = ['Year', 'Winner', 'Loser'])

postseason['Year'] = postseason['Year'].str.extract('(\d{4})+').astype('int')
postseason['Winner'] = postseason['Winner'].str.extract('([\w\s]+\.?[\w\s]+)\*? \(')
postseason['Loser'] = postseason['Loser'].str.extract('([\w\s]+\.?[\w\s]+)\*? \(')

postseason

    
```

```python
pitchData['playoff_status'] = range(0,len(pitchData)) #placeholder
for year in range(2000, 2022):
    year_tm = pitchData.loc[pitchData['Year'] == year, 'Tm'].unique()
    playoff_team = postseason.loc[postseason['Year'] == year, ]

    playoff_team = pd.concat([playoff_team['Winner'], playoff_team['Loser']], axis = 0, ignore_index = True).unique()


    status = [1 if team in list(playoff_team) else 0 for team in list(year_tm)]
    #Substitute that year's data with playoff status for those teams
    pitchData.loc[pitchData['Year'] == year, 'playoff_status'] = status
```

```python
pitchData.to_csv("Advanced_Pitch_Data.csv", index = False)

```
